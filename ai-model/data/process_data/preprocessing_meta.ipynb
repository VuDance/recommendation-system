```python
js_string = "var aPageStart = (new Date()).getTime();"

mask = meta_df['title'].str.contains(js_string, na=False, regex=False)

dirty = meta_df[mask]
print(f"Tổng số dòng bị lỗi: {len(dirty)}\n")
print(dirty[['title']].head(20))

# Xóa các dòng đó
df_clean = meta_df[~mask].reset_index(drop=True)
print(f"\nSau khi xóa: {len(df_clean)} dòng (đã xóa {len(meta_df) - len(df_clean)} dòng)")

df_clean = df_clean.drop_duplicates(subset=['title'], keep='first').reset_index(drop=True)
```

```python
from pathlib import Path
import pandas as pd
import json
import ast
from sqlalchemy import create_engine, text
import time

print("🔄 Bắt đầu quá trình import...")

CSV_PATH = Path(r"D:\MyWorkspace\recommendation-system\ai-model\data\dataset\products_clean.csv")
DB_CONFIG = {
    "user": "postgres",
    "password": "password123",
    "host": "18.140.225.1",
    "port": "5432",
    "dbname": "recommendation_db"
}
TABLE_NAME = "products"

engine = create_engine(
    f"postgresql+psycopg2://{DB_CONFIG['user']}:{DB_CONFIG['password']}@"
    f"{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['dbname']}"
)

print("📂 Đang đọc file CSV...")
df = pd.read_csv(CSV_PATH, low_memory=False)

columns_to_keep = ['asin', 'title', 'description', 'brand', 'imageURLHighRes']
existing_cols = [c for c in columns_to_keep if c in df.columns]
df = df[existing_cols].copy()

def to_json_array(val):
    if pd.isna(val) or val == '':
        return '[]'
    try:
        parsed = ast.literal_eval(val)
        return json.dumps(parsed)
    except:
        return '[]'

df['image_url'] = df['imageURLHighRes'].apply(to_json_array)
df = df.drop(columns=['imageURLHighRes'])
df.index.name = 'train_id'

# ✅ Tạo table với column image_url kiểu jsonb thủ công
with engine.connect() as conn:
    conn.execute(text(f"DROP TABLE IF EXISTS {TABLE_NAME}"))
    conn.execute(text(f"""
        CREATE TABLE {TABLE_NAME} (
            train_id SERIAL PRIMARY KEY,
            asin TEXT,
            title TEXT,
            description TEXT,
            brand TEXT,
            image_url JSONB
        )
    """))
    conn.commit()

# ✅ Insert từng batch với cast sang jsonb
start_time = time.time()
try:
    records = df.reset_index().to_dict(orient='records')
    with engine.connect() as conn:
        for i in range(0, len(records), 10000):
            batch = records[i:i+10000]
            conn.execute(
                text(f"""
                    INSERT INTO {TABLE_NAME} (train_id, asin, title, description, brand, image_url)
                    VALUES (:train_id, :asin, :title, :description, :brand, CAST(:image_url AS jsonb))
                """),
                batch
            )
            conn.commit()
            print(f"✅ Inserted {min(i+10000, len(records))}/{len(records)}")

    print(f"🎉 HOÀN THÀNH trong {time.time() - start_time:.2f} giây!")
except Exception as e:
    print(f"❌ Lỗi: {e}")
```
