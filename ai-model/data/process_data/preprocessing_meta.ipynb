```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
%load_ext autoreload
%autoreload 2


meta_df = pd.read_json("D:/MyWorkspace/recommendation-system/ai-model/data/dataset/meta_AMAZON_FASHION.json.gz", lines=True, compression="gzip")
meta_df.head()
print(len(meta_df))
```

```python
duplicate_counts = meta_df.groupby('title').size().reset_index(name='count')
duplicate_counts = duplicate_counts[duplicate_counts['count'] > 1].sort_values('count', ascending=False)

print(f"Tổng số title bị duplicate: {len(duplicate_counts)}\n")
print(duplicate_counts)
```

```python
js_string = "var aPageStart = (new Date()).getTime();"

mask = meta_df['title'].str.contains(js_string, na=False, regex=False)

dirty = meta_df[mask]
print(f"Tổng số dòng bị lỗi: {len(dirty)}\n")
print(dirty[['title']].head(20))

# Xóa các dòng đó
df_clean = meta_df[~mask].reset_index(drop=True)
print(f"\nSau khi xóa: {len(df_clean)} dòng (đã xóa {len(meta_df) - len(df_clean)} dòng)")

df_clean = df_clean.drop_duplicates(subset=['title'], keep='first').reset_index(drop=True)
```

```python
print(len(df_clean))
```

```python
duplicate_counts = df_clean.groupby('title').size().reset_index(name='count')
duplicate_counts = duplicate_counts[duplicate_counts['count'] > 1].sort_values('count', ascending=False)

print(f"Tổng số title bị duplicate: {len(duplicate_counts)}\n")
print(duplicate_counts)
```

```python
from pathlib import Path
import pandas as pd
from sqlalchemy import create_engine
import time

print("🔄 Bắt đầu quá trình import...")

# ====================== CẤU HÌNH ======================
CSV_PATH = Path(r"D:\MyWorkspace\recommendation-system\ai-model\data\dataset\products_clean.csv")
DB_CONFIG = {
    "user": "postgres",
    "password": "password123",
    "host": "localhost",
    "port": "5432",
    "dbname": "recommendation_db"
}
TABLE_NAME = "products"

engine = create_engine(
    f"postgresql+psycopg2://{DB_CONFIG['user']}:{DB_CONFIG['password']}@"
    f"{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['dbname']}"
)

# ====================== ĐỌC FILE CSV ======================
print("📂 Đang đọc file CSV...")

# Đọc bình thường, KHÔNG dùng index_col
df = pd.read_csv(CSV_PATH, low_memory=False)

print(f"Cột thực tế: {df.columns.tolist()}")

# Chỉ giữ các cột cần thiết
columns_to_keep = ['asin', 'title', 'description', 'brand', 'imageURL']
existing_cols = [c for c in columns_to_keep if c in df.columns]
df = df[existing_cols].copy()

# Đổi tên imageURL -> image_url cho khớp với Java Entity @Column(name = "image_url")
df = df.rename(columns={'imageURL': 'image_url'})

# Tự tạo train_id là số nguyên tăng dần từ 0
df.index.name = 'train_id'

print(f"✅ Đọc xong! Shape: {df.shape}")
print(f"Các cột: {df.columns.tolist()}")
print("📊 Kiểm tra 5 dòng đầu:")
print(df.head())

# ====================== IMPORT VÀO POSTGRESQL ======================
start_time = time.time()
try:
    df.to_sql(
        name=TABLE_NAME,
        con=engine,
        if_exists='replace',
        index=True,
        index_label='train_id',
        chunksize=10000,
        method='multi'
    )
    print(f"🎉 HOÀN THÀNH trong {time.time() - start_time:.2f} giây!")
except Exception as e:
    print(f"❌ Lỗi: {e}")
```
